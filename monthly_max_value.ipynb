```python
import pandas as pd
import numpy as np
```

```python
def monthly_max_value(file):
    df = pd.read_csv(file)
    df.drop(columns=['Open', 'High', 'Low', 'Adj Close', 'Volume'], inplace=True)
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)
    fridays = df.resample('W-FRI').apply(lambda x: x[-1])
    values = fridays[fridays['Close'] == fridays.groupby([fridays.index.year, fridays.index.month])['Close'].transform('max')]
    return values
```

```python
monthly_max_value('AAPL.csv')
```

```python
monthly_max_value('MSFT.csv')
```
